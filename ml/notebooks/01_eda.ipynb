# EDA — E-Commerce Fraud Detection

Exploratory analysis of synthetic transaction dataset.

```python
import sys
sys.path.insert(0, '../backend')

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from app.ml.features import build_training_dataframe

df = build_training_dataframe(n_samples=20000)
df.head()
```

```python
print('Fraud rate:', df['is_fraud'].mean())
print(df['is_fraud'].value_counts())
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='log_transaction_amount', hue='is_fraud', bins=40, kde=True)
plt.title('Transaction Amount by Fraud Label')
plt.show()
```

```python
numeric_cols = [c for c in df.columns if c != 'is_fraud']
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols + ['is_fraud']].corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```
